```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("happyscore_income.csv")
```

```python
df.groupby(by="region").mean()
```

```python
df2 = df.groupby(by="region").mean()
```

```python
df2
```

```python
plt.figure(figsize=(30,20))
sns.barplot(x=df2.index, y=df2.happyScore, data=df2)
plt.title("Mean Happy Score by Region", size=20)
plt.xlabel("Region", size=20)
plt.ylabel("Mean Happy Score", size=20)

plt.show()
```
